```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load file
dataset = pd.read_csv('C:/Users/ASUS/Documents/AA/Activity2/Iris.csv')
```

```python
# Show all the lines on the file
dataset.head(150)
```

```python
# label of the variables
dataset['Species'] = dataset['Species'].replace('Iris-setosa',0)
dataset['Species'] = dataset['Species'].replace('Iris-versicolor',1)
dataset['Species'] = dataset['Species'].replace('Iris-virginica',2)
```

```python
# Prints the dataset
dataset
```

```python
# Lenght of the dataset
dataset.shape
```

```python
# Isolate the variable of interest 
x = dataset.drop('Species', axis=1)
y = dataset['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101, shuffle= True, stratify=y)
# The test size is 30%, 
# The other parameters ensure that the algorithm doesn't focus just on one group
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
# Adjust the model to the data 
```

```python
predictions = logmodel.predict(x_test)
print('Predictions:\n', predictions)
# Make predictions
```

```python
from sklearn.metrics import confusion_matrix
results = confusion_matrix(y_test, predictions)
print('Confusion Matrix:\n', results)
# Confusion Matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
# The classification report visualizer displays the precision and support scores for the model.
```

```python
from sklearn import metrics
print('Accuracy:\n', metrics.accuracy_score(y_test,predictions))
# Show how accurate is our prediction
```
