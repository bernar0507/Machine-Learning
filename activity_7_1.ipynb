```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Utilizador/Desktop/Bioinf/2ยบ ano/2ยบ semestre/AA/Naive-Bayes-master/Naive-Bayes-Classification-Data.csv")
df.head()
```

```python
x = df.drop('diabetes', axis=1)
y = df['diabetes']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
model = GaussianNB()

# Fit Gaussian Naive Bayes according to X, y
model.fit(x_train, y_train)
```

```python
# Get parameters for this estimator
model.get_params()
```

```python
# Incremental fit on a batch of samples
model.partial_fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_test
```

```python
# Returns the log-probability of the samples for each class in the model
model.predict_log_proba(x_test)
```

```python
# Returns the probability of the samples for each class in the model
model.predict_proba(x_test)
```

```python
# Predict the response for the test datasets
y_pred = model.predict(x_test)
```

```python
# Compute confusion matrix to evaluate the accuracy of a classification
print(metrics.confusion_matrix(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)*100
accuracy
```
